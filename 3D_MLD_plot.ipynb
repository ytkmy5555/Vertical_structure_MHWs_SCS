```python
def rotate_90_degrees(matrix):
    n = len(matrix)
    result = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            result[n-j-1][i] = matrix[i][j]
    
    return result
rotated_matrix = rotate_90_degrees(mld1[0])
```
